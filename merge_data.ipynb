```python
import pandas as pd

full_df = pd.read_csv('data/all_runs.csv', delimiter='\t')
full_df=full_df[full_df['Induction']==1]

# Round a columns to integer
full_df['Temperature'] = full_df['Temperature'].astype(int)
full_df.sort_values(by=['Temperature'], inplace=True)   

unique_temps = full_df['Temperature'].unique()

for temp in unique_temps:
    print(f'Temperature: {temp}')
    temp_df = full_df[full_df['Temperature']==temp]
    display(temp_df)
```
